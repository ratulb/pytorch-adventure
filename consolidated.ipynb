```python
import torch
import torch.nn as nn
import numpy as np
from sklearn import datasets
import matplotlib.pyplot as plt
```

```python
X_numpy, y_numpy = datasets.make_regression(n_samples=100,n_features=1, noise=10, random_state=2)

X = torch.from_numpy(X_numpy.astype(np.float32))
y = torch.from_numpy(y_numpy.astype(np.float32))
y = y.view(y.shape[0],1)
n_samples, n_features = X.shape
input_size = n_features
output_size = n_features
model = nn.Linear(input_size, output_size)
criterion = nn.MSELoss()
learning_rate = 0.01
optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)
```

```python
n_iter = 1000
for epoch in range(n_iter):
    y_predicted = model(X)
    loss = criterion(y_predicted, y)
    loss.backward()
    optimizer.step()
    optimizer.zero_grad()

    if (epoch+1) % 10 == 0:
        [w,b] = model.parameters()
        print(f'epoch = {epoch+1}, weight = {w[0][0].item():.3f}, loss = {loss.item():.3f}')


predicated = model(X).detach().numpy()
plt.plot(X_numpy, y_numpy, 'ro')
plt.plot(X_numpy, predicated, 'b')
plt.show()
```
